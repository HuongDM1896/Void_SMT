```python
# Cell 1: Import libs
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Cell 2: Import image and threshold
image = cv2.imread('4.png', cv2.IMREAD_GRAYSCALE)

# Hin th nh gc
plt.imshow(image, cmap='gray')
plt.title('Ori')
plt.axis('off')
plt.show()

#Set threshold
_, threshold = cv2.threshold(image, 70, 255, cv2.THRESH_BINARY)

# Show image
plt.imshow(threshold, cmap='gray')
plt.title('Correct')
plt.axis('off')
plt.show()
```

```python
# Cell 3: Calculate %
# S of Light and dark area 
total_area = image.shape[0] * image.shape[1]
bright_area = np.sum(threshold == 255)

# Light percent = Void percent
bright_percentage = (bright_area / total_area) * 100

# Show result
print(f"Void: {bright_area} pixels")
print(f"Fine: {total_area - bright_area} pixels")
print(f"Percent void: {bright_percentage:.2f}%")

# Check with limit and return "OK" or "FAIL"
result = "OK" if bright_percentage < 10 else "FAIL"
print(f"Result: {result}")
```
